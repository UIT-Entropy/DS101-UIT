Câu 1: Thực hiện mô phỏng hàm mật độ của ít nhất 3 phân phối (chẳng hạn: nhị thức, Poisson, Gauss, mũ,...)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python
#nhị thức
n, p = 100, 0.3
samples = binom.rvs(n, p, size=1000)
x = np.arange(0,n+1)
plt.plot(x, binom.pmf(x,n,p), label='Hàm mật độ xác suất của phân phối nhị thức')
```

```python
#poisson
lamda = 3
samples = poisson.rvs(lamda, size = 100)
x= np.arange(0,101)
plt.plot(x, poisson.pmf(x,lamda), label='Hàm mật độ xác suất của phân phối poisson')
```

```python
#Phân phối chuẩn
nuy, sigma = 0, 1
x = np.linspace(-1, 1, 200)
samples = norm.rvs(nuy, sigma, size=1000)
plt.plot(x, norm.pdf(x, nuy, sigma), label='Hàm mật độ xác suất của phân phối chuẩn')
```

Câu 6: Cho xích Markov có ma trận chuyển:
\[
P =
\begin{bmatrix}
0.8 & 0.2 \\
0.1 & 0.9
\end{bmatrix}


Hãy viết chương trình Python để tìm phân phối dừng của Xích Markov và cho biết sau bao lâu thì xích Markov này đạt phân phối dừng.




```python
p = np.array([[0.8, 0.2], [0.1, 0.9]])
a = np.transpose(p) - np.eye(2)
a = np.vstack([a, np.ones(2)])
b = np.array([0, 0, 1])
pi = np.linalg.lstsq(a, b)[0]
print("Phân phối dừng π =", pi)

pi0 = np.array([1.0, 0])
for t in range (1, 50):
  pit = np.dot(pi0, np.linalg.matrix_power(p,t))
  diff = np.linalg.norm(pit - pi)
  if(diff < 1e-6):
    print(f"xích markov đạt phân phối dừng trong khoảng {t} bước, với sai khác của phân phối là {diff}")
    break
```

\[
P =
\begin{bmatrix}
0.64 & 0.32 & 0.04 \\
0.40 & 0.50 & 0.10 \\
0.25 & 0.50 & 0.25
\end{bmatrix}


Tìm phân phối dừng của xích Markov này. Hãy cho biết sau bao lâu thì xích Markov này đạt phân phối dừng.




```python
p = np.array([[0.64, 0.32, 0.04], [0.4, 0.5, 0.1], [0.25, 0.5, 0.25]])
a = np.transpose(p) - np.eye(3)
a = np.vstack([a, np.ones(3)])
b = np.array([0, 0, 0, 1])
pi = np.linalg.lstsq(a, b)[0]
print("Phân phối dừng π =", pi)
pi0 = np.array([1.0, 0, 0])
for t in range(1,200):
  pit = np.dot(pi0, np.linalg.matrix_power(p,t))
  diff = np.linalg.norm(pit - pi)
  if diff < 1e-6:
    print(f"xích markov đạt phân phối dừng trong khoảng {t} bước, với sai khác của phân phối là {diff}")
    break
```

Câu 8: Mô phỏng quá trình bước ngẫu nhiên đơn giản

Xét quá trình bước ngẫu nhiên đơn giản được định nghĩa như sau:

$$
X(t) = \sum_{i=1}^{t} Y_i, \qquad X(0) = 0
$$

trong đó

$$
Y_i =
\begin{cases}
+1, & \text{với xác suất } p, \\[6pt]
-1, & \text{với xác suất } 1 - p.
\end{cases}
$$

a) Mô phỏng quá trình khi p = 1/2 \
b) p là giá trị cho trước, vẽ 3 quỹ đạo 1/2, p, 1-p trên cùng 1 đồ thị


```python
#Câu a
def SumArr(n, p):
  x = uniform(0,1)
  y = x.rvs(size = n, random_state = 42)
  X = 0
  Sum = [X]
  for i in range(n):
    if y[i] < p:
      X = X + 1
    else:
      X = X - 1
    Sum.append(X)
  return Sum
t = np.arange(0, 1001)
plt.plot(t, SumArr(1000, 0.5), label = "p = 0.5")
```

```python
#Câu b
p = 0.8   #đổi p tùy chọn trong khoảng [0,1]
plt.figure(figsize=(10, 6))
plt.plot(t, SumArr(1000, 0.5), label = "p = 0.5")
plt.plot(t, SumArr(1000, p), label = f"p = {p}")
plt.plot(t, SumArr(1000, 1-p), label = f"p = {1-p}")
plt.legend()
plt.show()
```
